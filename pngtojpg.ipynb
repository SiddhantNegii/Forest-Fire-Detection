```python
import os
from PIL import Image

# Directory containing PNG files
input_dir = r'D:\Project_ResearchPaper\Fire_detection\Datasetfiredetection\Fire_data\images\train'
# Directory to save JPG files
output_dir = r'D:\Project_ResearchPaper\Fire_detection\Datasetfiredetection\train\jpg_images'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Check if input directory exists
if not os.path.exists(input_dir):
    print(f"Input directory does not exist: {input_dir}")
    exit(1)

print(f"Converting files from {input_dir} to {output_dir}")

# List all files in the input directory
input_files = os.listdir(input_dir)
print(f"Files in input directory: {input_files}")

# Loop through all files in the input directory
for filename in input_files:
    if filename.endswith(".png"):
        try:
            input_path = os.path.join(input_dir, filename)
            output_path = os.path.join(output_dir, os.path.splitext(filename)[0] + '.jpg')
            
            # Open the image file
            with Image.open(input_path) as img:
                # Convert to RGB (to remove alpha channel)
                rgb_img = img.convert('RGB')
                # Save as JPG
                rgb_img.save(output_path, 'JPEG')
            
            print(f"Converted {input_path} to {output_path}")

        except Exception as e:
            print(f"Failed to convert {filename}: {e}")

# List all files in the output directory
output_files = os.listdir(output_dir)
print(f"Files in output directory: {output_files}")

print("Conversion completed!")
```
